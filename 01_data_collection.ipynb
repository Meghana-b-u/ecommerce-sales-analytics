```python
import pandas as pd
import numpy as np

# Replace 'your_file.csv' with the name from Step 8
df = pd.read_csv(f'{BASE}/data/raw/online_retail_II.csv',
                 encoding='utf-8',
                 on_bad_lines='skip')

print(f"Rows    : {len(df):,}")
print(f"Columns : {list(df.columns)}")
```

```python
df.columns = df.columns.str.strip()

rename_map = {
    'InvoiceNo'  : 'Invoice',
    'UnitPrice'  : 'Price',
    'CustomerID' : 'Customer ID'
}
df.rename(columns=rename_map, inplace=True)

print("Columns after standardizing:")
print(list(df.columns))
```

```python
df['TotalRevenue'] = df['Quantity'] * df['Price']

print("=== DATA PROFILE ===")
print(f"Rows          : {len(df):,}")
print(f"Columns       : {df.shape[1]}")
print(f"Duplicates    : {df.duplicated().sum():,}")
print(f"\nMissing values:")
print(df.isnull().sum()[df.isnull().sum() > 0])
print(f"\nGross revenue : £{df['TotalRevenue'].sum():,.0f}")
print(f"Return rows   : {(df['TotalRevenue']<0).sum():,}")
```

```python
df.to_csv(f'{BASE}/data/processed/stage1_loaded.csv',
              index=False)

print("Stage 1 DONE!")
print("File saved to Drive.")
print("\nNext stage starts with just this line:")
print("  df = pd.read_csv(f'{BASE}/data/processed/stage1_loaded.csv')")
```
